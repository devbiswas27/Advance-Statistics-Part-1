Q18.What is the significance of confidence intervals in statistical analysis?

A confidence interval (CI) is a range of values used to estimate a population parameter (like a mean or proportion), with an associated level of confidence (like 95%). It adds context and reliability to point estimates, making it a core tool in inferential statistics.

 Why Confidence Intervals Matter:

 1. Quantifies Uncertainty

A point estimate (e.g., sample mean) gives one value ‚Äî but a confidence interval shows how much uncertainty is in that estimate.

It answers: ‚ÄúHow precise is our estimate?‚Äù

 2. Provides a Range for the True Parameter

A 95% CI means: ‚ÄúWe are 95% confident that the true population value lies within this range.‚Äù

It does not mean there's a 95% chance the true value is in the interval ‚Äî the interval either does or doesn't contain it; the 95% refers to the method's reliability over many samples.

 3. Supports Decision Making

Wider intervals ‚Üí more uncertainty

Narrower intervals ‚Üí more precise estimate
Helps you decide whether a treatment effect, business outcome, or policy impact is statistically meaningful.

 4. Preferred Over P-values Alone

Confidence intervals show the effect size and its range, while a p-value just tells you whether something is statistically significant.

Example: A p-value might say ‚Äúsignificant,‚Äù but a CI shows how large or small the effect might actually be.

 5. Used in Hypothesis Testin
If the null hypothesis value (e.g.
Œº=0) falls outside the confidence interval, you can often conclude that the result is statistically significant at the chosen level (like 5%).

Q23.How does a probability distribution relate to the expected outcome of a random variable?

 Relationship in Simple Terms:

The probability distribution tells you what could happen and how likely each outcome is.
